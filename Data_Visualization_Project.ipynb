```python
codes = pd.read_csv('/content/drive/MyDrive/MyProject2/isocodes.csv')
codes = codes.rename(columns={'English short name lower case': 'Area'})
nat_fire = pd.read_csv('/content/drive/MyDrive/MyProject2/NatFires.csv')
area_burned = nat_fire.query("`Element` == 'Burned Area'").groupby(['Area', 'Year'])['Value'].sum().reset_index()
area_burned.iloc[1076:1081, 0] = pd.Series(['Turkey'] * 5)
codes.iloc[232, 0] = 'United States of America'
codes.iloc[238, 0] = 'Viet Nam'
codes.iloc[115, 0] = "Democratic People's Republic of Korea"
area_burned.iloc[224:229, 0] = pd.Series(['Taiwan'] * 5)
codes.iloc[116, 0] = "Republic of Korea"
codes.iloc[116, 0] = "Republic of Korea"
codes.iloc[57, 0] = 'Czechia'
codes.iloc[143, 0] = 'Republic of Moldova'
codes.iloc[216, 0] = 'United Republic of Tanzania'

# This function takes two tables and their factoral column name and checks if values of first table`s column
# are substring of second and if values of second table`s column are substring of first. If yes, it changes value
# in the second table so that it would have analogical in first. It is gonna be used many times as we are working with
# different tables containing country names, which do not correspond always.

def equal_vals(tab1, tab2, col: str) -> None:
    try:
        vals1 = list(tab1[col])
        vals2 = list(tab2[col])
    except KeyError:
        raise KeyError('Column is not in table.')
    else:
        nocorr = set()
        for k, i in enumerate(vals2):
            if not isinstance(i, str):
                raise TypeError('Function is only applied to factoral columns')
            if i not in vals1:
                for j in vals1:
                    if not isinstance(j, str):
                        raise TypeError('Function is only applied to factoral columns')
                    if (i in j) or (j in i):
                        vals2[k] = j
                        break
                else:
                    nocorr.add(i)

        tab2[col] = pd.Series(vals2)
        print(f'No corresponding for some reason have: \n{nocorr}') # this function may not work properly always, but it prints the values which were not
                                                                  #changed so we could know what values are needed to be changed in code soon.


#This function takes a data frame , it`s column name, apropriate year and label to create a choropleth
#using also additional data frames with country codes etc. Make sure that column containing country
#names in all additional tables is named 'Area' before using.

def create_choropleth(data_frame, column: str, year: int, label: str, *adds) -> None:
    data = data_frame.query(f"`Year` == {year}")
    for i in adds:
        data = data.merge(i, on='Area')
    fig = px.choropleth(data, color=data[column], locations=data['Alpha-3 code'], labels={column: label}, hover_name=column, projection = "mollweide")
    fig.show()

def transform_to_percentage(data_frame, trans_column: str, new_col_name: str = 'new'): pass
equal_vals(area_burned, codes, 'Area')
create_choropleth(area_burned, 'Value', 2001, 'Area burned in 2001', codes)
```

```python
def generalize(table, by: str, title: str, *countries):
    frame = table.copy(deep=True).query(f"Area in {str(list(countries))}")
    new_plot = px.line(data_frame=frame, x='Year', y=by, markers=True, color='Area')
    new_plot.show()

generalize(area_burned, 'Value', 'Area burned each year', *list(area_burned.copy(deep=True).sort_values(by='Value', ascending=False).iloc[0:38, 0]))


```
