```python
# The table that describes which type of coverage is typical or not for each continent in 2001-2021. Soon will help us to understand the connection between climat and the
# fenomenon of natural fires.
world_c1 = pd.read_csv('/content/drive/MyDrive/MyProject/WorldCover1.csv')
# Some of data rows in column 'Item' here are empty everywhere, so we can delete ones.
world_c1.dropna(inplace=True, axis=0)
# The graph which describes natural plant coverage in the world in general in 2001-2021.
world_c1_gen = world_c1.query("`Area` == 'World'")
fig1_c1 = px.bar(data_frame=world_c1_gen, x='Value', y='Item', animation_frame='Year', animation_group='Item', color='Item', range_x=[0, 6e6])
fig1_c1.show()
# The values have not changed that much in 20 years. It is seen that Earth is mostly covered with forests and grasslands. And we can say that they occur in most of climatic zones.
```

```python
world_c2 = pd.read_csv('/content/drive/MyDrive/MyProject/all_burned_areas_ha.csv')
fig2_c1 = px.bar(data_frame=world_c2, x='Value', y='Item', animation_frame='Year', animation_group='Item', color='Item', range_x=[0, 1.8e8])
fig2_c1.show()
```
